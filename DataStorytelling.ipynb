```python
# Import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# https://www.kaggle.com/datasets/andrewmvd/data-scientist-jobs?resource=download
# Load downloaded data
data = pd.read_csv('DataScientist.csv')
```

```python
# drop useless columns
data = data.drop(['Unnamed: 0', 'index', 'Company Name', 'Headquarters', 'Type of ownership', 'Founded', 'Competitors','Revenue'], axis=1)
```

```python
# All missing values are shown as'-1' in the dataset
# Count how many '-1' in each column
# Create an empty dictionary to store the counts for each column
counts = {}

# Loop through the columns and count '-1' values in each column
for column in data.columns:
    count_minus_one = (data[column] == '-1').sum() + (data[column] == -1).sum()
    counts[column] = count_minus_one

# Print the counts for each column
for column, count in counts.items():
    print(f"Number of '-1' values in '{column}': {count}")
```

```python
# Create a filter for rows values equals '-1'
filter_condition = (data['Industry'] == '-1') | (data['Rating'] == -1)

# Use the filter to drop the rows from the DataFrame
data = data[~filter_condition]
```

```python
# Check unique values of column 'Easy Apply'
data['Easy Apply'].unique()
```

```python
# Update value '-1' with 'False'
data['Easy Apply'] = data['Easy Apply'].replace('-1', False)
data['Easy Apply'] = data['Easy Apply'].replace('True', True)
data['Easy Apply'].unique()
```

```python
# Recount '-1'
counts = {}

# Loop through the columns and count '-1' values in each column
for column in data.columns:
    count_minus_one = (data[column] == '-1').sum() + (data[column] == -1).sum()
    counts[column] = count_minus_one

# Print the counts for each column
for column, count in counts.items():
    print(f"Number of '-1' values in '{column}': {count}")
```

```python
data['Salary Estimate'].value_counts()
```

```python
import re 

# Create two new columns 'Min Salary(K)' and 'Max Salary(K)'
data['Min Salary(K)'] = 0
data['Max Salary(K)'] = 0

for index, row in data.iterrows():
    salary_estimate = row['Salary Estimate']
    
    if 'Per Hour' in salary_estimate:
        # Extract the numeric part of the salary estimate using regular expressions
        salary_values = re.findall(r'\d+', salary_estimate)
        min_salary, max_salary = map(int, salary_values)
        data.at[index, 'Min Salary(K)'] = min_salary * 2080
        data.at[index, 'Max Salary(K)'] = max_salary * 2080
    else:
        # Extract the numeric part of the salary estimate using regular expressions
        salary_values = re.findall(r'\d+', salary_estimate)
        min_salary, max_salary = map(int, salary_values)
        data.at[index, 'Min Salary(K)'] = min_salary
        data.at[index, 'Max Salary(K)'] = max_salary

# Drop the original 'Salary Estimate' column
data = data.drop(columns='Salary Estimate')
```

```python
data['Max Salary(K)'].value_counts().sort_index(ascending=False)
```

```python
values_to_show = [70720, 35360, 20800]  # Replace with the values you want to filter

# Create a boolean mask for the specified values in the 'Min Salary(K)' column
mask = data['Min Salary(K)'].isin(values_to_show)

# Use the mask to filter and replace values
data.loc[mask, 'Min Salary(K)'] = round(data.loc[mask, 'Min Salary(K)'] / 1000)
data.loc[mask, 'Max Salary(K)'] = round(data.loc[mask, 'Max Salary(K)'] / 1000)
```

```python
data['Location'].value_counts()
```

```python
# Split the 'Location' column into 'City' and 'State' columns
data[['City', 'State']] = data['Location'].str.split(', ', n=1, expand=True)

# Drop the original 'Location' column if you no longer need it
data = data.drop(columns='Location')
```

```python
data['Size'].value_counts()
```

```python
# Filter rows where 'Size' is not equal to 'Unknown'
data = data[data['Size'] != 'Unknown']

# Reset the index if needed
data = data.reset_index(drop=True)

# Remove ' employees' from every value in the 'Size' column
data['Size'] = data['Size'].str.replace(' employees', '')

# Rename the 'Size' column to 'Employees'
data = data.rename(columns={'Size': 'Employees'})
```

```python
data['Job Title'].value_counts()
```

```python
# Define conditions and corresponding values for the 'Job Level' column
conditions = [
    data['Job Title'].str.contains('Data Analyst', case=False),
    data['Job Title'].str.contains('Data Engineer', case=False),
    data['Job Title'].str.contains('Data Scientist', case=False),
    data['Job Title'].str.contains('Researcher', case=False),
    data['Job Title'].str.contains('Consultant', case=False),
    data['Job Title'].str.contains('Data Science', case=False),
]
choices = ['Data Analyst', 'Data Engineer', 'Data Scientist', 'Researcher', 'Consultant', 'Other Data Science']

# Create the 'Job Level' column based on the conditions and choices
data['Job Level'] = np.select(conditions, choices, default='Other Positions')
```

## Where are the 3286 Data Science Jobs located?

```python
# Calculate the value counts by 'State'
state_counts = data['State'].value_counts().sort_values(ascending=False)

# Calculate the average minimum salary by 'State'
average_min_salary_by_state = data.groupby('State')['Min Salary(K)'].mean()

# Calculate the overall average minimum salary
overall_average_min_salary = data['Min Salary(K)'].mean()

# Create a figure and primary Y-axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot the value counts on the primary Y-axis
state_counts.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_xlabel('State')
ax1.set_ylabel('Number of Job Listings', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Create a secondary Y-axis
ax2 = ax1.twinx()

# Plot the average minimum salary on the secondary Y-axis
ax2.plot(state_counts.index, average_min_salary_by_state[state_counts.index], color='orange', marker='o')
ax2.set_ylabel('Average Minimum Salary (in $K)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Set the secondary (right) y-axis limits to start from 0
ax2.set_ylim(0)

# Add a horizontal line for the overall average minimum salary
plt.axhline(overall_average_min_salary, color='red', linestyle='--', label='Overall Average', alpha=0.7)

# Annotate the line with the average number (above the line)
plt.text(state_counts.index[-1], overall_average_min_salary + 2, f'Avg Min Salary: {overall_average_min_salary:.2f}', color='red', fontsize=12, horizontalalignment='right')

# Set labels and title
plt.title('Data Science Job Listings and Average Minimum Salary by State')
ax1.set_xticklabels(state_counts.index, rotation=0)

# Add a legend
ax2.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

### California Average Minimum Salary: 105.74K, 28.06% higher than national average 82.57K.

```python
# Calculate the total average minimum salary for all states
total_average_min_salary_all_states = data['Min Salary(K)'].mean()

# Calculate the average minimum salary for California
average_min_salary_california = data[data['State'] == 'CA']['Min Salary(K)'].mean()

# Calculate the percentage difference
percentage_difference = ((average_min_salary_california - total_average_min_salary_all_states) / total_average_min_salary_all_states) * 100

print(f"Total Average Minimum Salary for All States: ${total_average_min_salary_all_states:.2f}K")
print(f"Average Minimum Salary for California (CA): ${average_min_salary_california:.2f}K")
print(f"California (CA) is {percentage_difference:.2f}% higher than the average for all states.")
```

## Data Science Job Market in California

```python
# Define the lists of cities for NorCal and SoCal
norcal_cities = ['San Jose', 'Santa Clara', 'Mountain View', 'Redwood City', 'Palo Alto', 'Sunnyvale', 'Menlo Park', 'La Jolla', 'El Segundo', 'Fremont', 'Pasadena', 'Cupertino', 'Irwindale', 'Newark', 'Livermore', 'Pleasanton', 'San Carlos', 'Monrovia', 'Northridge', 'Los Gatos', 'Campbell', 'Poway', 'Cerritos', 'San Mateo', 'Milpitas']
socal_cities = ['San Diego', 'Los Angeles', 'Santa Monica', 'Burbank', 'Culver City', 'Glendale', 'Long Beach', 'Irvine', 'Torrance', 'Woodland Hills', 'Marina del Rey', 'Sherman Oaks', 'Whittier', 'Calabasas', 'Universal City', 'Anaheim', 'Brea', 'Westmont', 'Venice', 'Wilmington', 'Chula Vista', 'Seal Beach']

# Filter the data for NorCal and SoCal
norcal_data = data[data['City'].isin(norcal_cities)]
socal_data = data[data['City'].isin(socal_cities)]

# Calculate the counts for NorCal and SoCal
norcal_count = norcal_data.shape[0]
socal_count = socal_data.shape[0]

# Calculate the average minimum salary for NorCal and SoCal
norcal_avg_salary = norcal_data['Min Salary(K)'].mean()
socal_avg_salary = socal_data['Min Salary(K)'].mean()

# Create a DataFrame for the counts and salaries
data_dict = {
    'Region': ['North California', 'South California'],
    'Count': [norcal_count, socal_count],
    'Average Min Salary (K)': [norcal_avg_salary, socal_avg_salary]
}
counts_df = pd.DataFrame(data_dict)

# Create a stacked bar chart with two Y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the counts on the left Y-axis
counts_df.plot(kind='bar', x='Region', y='Count', stacked=True, ax=ax1, color='skyblue', legend=False)
ax1.set_xlabel('Region')
ax1.set_ylabel('Count', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Create a secondary Y-axis for salary
ax2 = ax1.twinx()
ax2.plot(counts_df['Region'], counts_df['Average Min Salary (K)'], marker='o', color='orange', label='Avg Min Salary (K)')
ax2.set_ylabel('Average Min Salary (K)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Set the right Y-axis (salary) to start from 0
ax2.set_ylim(bottom=0)

# Set the x-axis labels to be horizontal
ax1.set_xticklabels(counts_df['Region'], rotation=0)

# Show the plot with a legend
plt.title('Job Listings Counts and Average Min Salary in NorCal and SoCal')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

## How do job levels or titles affect salaries?

```python
# Calculate the average minimum salary for each job level
average_min_salary_by_job_level = data.groupby('Job Level')['Min Salary(K)'].mean()

# Sort the job levels by average minimum salary in descending order
sorted_job_levels = average_min_salary_by_job_level.sort_values(ascending=False).index

plt.figure(figsize=(10, 6))

# Create a bar chart for the average maximum salary with a different color
ax = sns.barplot(x='Job Level', y='Max Salary(K)', data=data, errorbar=None, color='orange', label='Avg Max Salary', order=sorted_job_levels)

# Create a bar chart to show the average minimum salary for each job level
ax = sns.barplot(x='Job Level', y='Min Salary(K)', data=data, errorbar=None, color='skyblue', label='Avg Min Salary', order=sorted_job_levels)

plt.xlabel('Job Level')
plt.ylabel('Average Salary (in $K)')
plt.title('Relationship between Job Level and Average Salary (Ranked by Minimum Salary)')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

# Add Y values above each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}K', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

# Add a legend for the average salary lines
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

### Minimum Average Salary for 'Data Scientist' is 97.60K, 18.20% higher than total average $82.57K.

```python
# Calculate the average minimum salary for all jobs
average_min_salary_all_jobs = data['Min Salary(K)'].mean()

# Filter the data for 'Data Scientist' job level
data_scientist_data = data[data['Job Level'] == 'Data Scientist']

# Calculate the minimum average salary for 'Data Scientist' job level
min_average_salary_data_scientist = data_scientist_data['Min Salary(K)'].mean()

# Calculate the percentage increase
percentage_increase = ((min_average_salary_data_scientist - average_min_salary_all_jobs) / average_min_salary_all_jobs) * 100

print(f"Average Minimum Salary for All Jobs: ${average_min_salary_all_jobs:.2f}K")
print(f"Minimum Average Salary for 'Data Scientist' Job Level: ${min_average_salary_data_scientist:.2f}K")
print(f"'Data Scientist' Job Level has an average minimum salary {percentage_increase:.2f}% higher than the average minimum salary for all jobs.")
```

## Does a higher salary mean a job is harder to get? 

```python
# Calculate the average minimum salary by job level
average_min_salary_by_job_level = data.groupby('Job Level')['Min Salary(K)'].mean()

# Calculate the "Easy Apply" percentage by job level
easy_apply_percentage_by_job_level = (data[data['Easy Apply']].groupby('Job Level').size() / data.groupby('Job Level').size()) * 100

# Calculate the total average "Easy Apply" percentage
total_average_easy_apply_percentage = (data[data['Easy Apply']].shape[0] / data.shape[0]) * 100

# Create a DataFrame for both values
job_level_data = pd.DataFrame({'Average Min Salary (K)': average_min_salary_by_job_level, 'Easy Apply Percentage (%)': easy_apply_percentage_by_job_level})

# Sort the DataFrame by average minimum salary in descending order
job_level_data = job_level_data.sort_values(by='Average Min Salary (K)', ascending=False)

plt.figure(figsize=(10, 6))

# Create a bar chart for the average minimum salary with a different color
ax = sns.barplot(x=job_level_data.index, y='Average Min Salary (K)', data=job_level_data, color='skyblue', label='Avg Min Salary')

# Set the right Y-axis
ax2 = ax.twinx()
ax2.plot(job_level_data.index, job_level_data['Easy Apply Percentage (%)'], marker='o', color='orange', label='Easy Apply Percentage')
ax2.set_ylabel('Easy Apply Percentage (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.xlabel('Job Level')
plt.ylabel('Average Min Salary (K) / Easy Apply Percentage (%)')
plt.title('Relationship between Job Level, Average Min Salary, and Easy Apply Percentage (Ranked by Salary)')

# Add a red line for the total average "Easy Apply" percentage
plt.axhline(total_average_easy_apply_percentage, color='red', linestyle='--', label='Total Avg Easy Apply Percentage', alpha=0.7)

# Add Y values above each bar for the average minimum salary
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}K', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

# Add a legend for both lines
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

# Set both Y-axes to start from 0
ax.set_ylim(0)
ax2.set_ylim(0)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
